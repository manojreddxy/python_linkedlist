```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# The 'error_bad_lines=False' argument will skip the problematic rows.
# 'quoting=3' will ignore the quotes
data = pd.read_csv('/content/fake_news.csv', on_bad_lines='skip', quoting=3)

data.head()
```

```python
data.isna().sum()
```

```python
data=data.drop(['id'],axis=1)
```

```python
data=data.fillna('')
```

```python
data['content'] = data['author']+' '+ data['title']+' '+data['text']




```

```python
data=data.drop(['title','author','text'],axis=1)
```

```python
data.head()
```

```python
# Convert to lowercase

data['content'] = data['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))




```

```python
# Remove punctuation

data['content'] = data['content'].str.replace('[^\w\s]','')




```

```python
# Remove stop words

import nltk

nltk.download('stopwords')




```

```python
from nltk.corpus import stopwords

stop = stopwords.words('english')

data['content'] = data['content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))




```

```python
# Do lemmatization
import nltk

# Download WordNet
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
from textblob import Word

data['content'] = data['content'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

data['content'].head()
```

```python
X=data[['content']]
y=data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting into training and testing data
# Check for classes with only one sample
from collections import Counter

class_counts = Counter(y)
rare_classes = [cls for cls, count in class_counts.items() if count < 2]

if rare_classes:
    print(f"Classes with less than 2 samples: {rare_classes}")
    # Remove rows with rare classes or handle them differently
    # For example, you can oversample them or remove them
    # Here, we remove them:
    data = data[~data['label'].isin(rare_classes)]
    X = data[['content']]
    y = data['label']

# Now you can safely use stratify
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45, stratify=y)
```

```python
#validate the shape of train and test dataset

print (X_train.shape)

print (y_train.shape)

print (X_test.shape)

print (y_test.shape)




```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)

tfidf_vect.fit(data['content'])

xtrain_tfidf = tfidf_vect.transform(X_train['content'])

xtest_tfidf = tfidf_vect.transform(X_test['content'])
```
